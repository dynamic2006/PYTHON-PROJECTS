```python
import cv2
import numpy as np
import time
import matplotlib.pyplot as plt

video_path = ''

def segment_video(video_path):
    import time
    import cv2
    import numpy as np
    import matplotlib.pyplot as plt

    # Start timing
    start_time = time.time()

    # Load video
    cap = cv2.VideoCapture(video_path)
    if not cap.isOpened():
        print("Error: Unable to open video.")
        return

    # Variables for storing ROI data
    roi_intensity_traces = {}
    frame_idx = 0

    # Process each frame
    while True:
        ret, frame = cap.read()
        if not ret:
            break  # End of video

        # Convert to grayscale and threshold
        gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        _, binary = cv2.threshold(gray, 128, 255, cv2.THRESH_BINARY)

        # Find connected components or contours
        num_labels, labels, stats, centroids = cv2.connectedComponentsWithStats(binary)

        # Extract intensity for each ROI
        for i in range(1, num_labels):  # Skip background (label 0)
            mask = (labels == i)
            mean_intensity = np.mean(gray[mask])

            # Initialize the key if it doesn't exist
            if i not in roi_intensity_traces:
                roi_intensity_traces[i] = []

            # Append mean intensity
            roi_intensity_traces[i].append(mean_intensity)

        # Increment frame index
        frame_idx += 1

    # Stop timing
    end_time = time.time()

    # Print performance
    print(f"Segmentation completed in {end_time - start_time:.2f} seconds")

    # Release video capture
    cap.release()

    # Display intensity traces
    plt.figure(figsize=(10, 6))
    for roi_id, trace in roi_intensity_traces.items():
        plt.plot(trace, label=f"ROI {roi_id}")
    plt.xlabel("Frame Index")
    plt.ylabel("Intensity")
    plt.title("Intensity Traces for Each ROI")
    plt.legend()
    plt.show()


# Replace with your video file path
video_path =  "/home/tejasvi/Downloads/video_data.avi"
segment_video(video_path)
```
